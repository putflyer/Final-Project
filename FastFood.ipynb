## The Metric: Fast Food Places

The second metric we looked at was fast food places. If we are going to deliver food to people we have to know of course where the fast food places are that people can order from. 

The delivery drivers who are bikers need to know which neighborhoods these fast food places are located. This can help us look for the best neighborhood as the more total fast food places as well as unique fast food places can really help determine which neighborhood or area is the best for a biker delivery doordasher.

This data set looks at all the fast food and major food places in Pittsburgh and groups them by zipcode. There are over 800 total restaurants on here. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fastfood_data = pd.read_csv('fastfooddata.csv', usecols=["Name", "ZIP Code"])

fastfood_data.plot.bar(x='Name', y="ZIP Code", figsize=(25,10))
```

As seen in our first set of data here are all the restaurants located with their zipcodes that are not sorted in any way. Basically though this chart seems really raw and impossible to read these are all the different retaurants even every franchise isnt separated as well as the zipcodes for each and every one. 

Now I am going to clump all the restaurants that follow the same zipcode together and sort them in an actual readable chart to see how many restaurants fall under each zipcode.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fastfood_data2 = pd.read_csv('fastfooddata2.csv', usecols=["ZIP Code", "Quantity"])
fastfood_data2.groupby(["ZIP Code"])["Quantity"].sum().plot.bar(legend=True , figsize = (15,10))

```

Based on this chart, we can see our true analysis for this metric. This was discovering how many restaurants can be found in each zipcode.
