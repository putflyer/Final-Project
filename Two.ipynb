# The Metric: Households

Our final dataset is households. If we're going to deliver food we are going to need to know where people who might order are. This data set lists households by race as well as by their status as a family, however everyone needs to eat, so we're going to ignore that information. For our purposes a single father with kids is going to order as much food as a college student living alone, we're delivery drivers so all that really matters is the tip we get at the end.

To start we're just going to lay out all of information as a whole and then look at the highs and the lows.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

houses = pd.read_csv('two.csv', usecols=["Estimate; Total:", "Neighborhood"])

houses.plot.bar(x='Neighborhood', y="Estimate; Total:", figsize=(15,5))
```

From what we can see there is a large spread of people throughout the city. Let's look at the outliers.

```python
houses = pd.read_csv('two.csv', usecols=["Estimate; Total:", "Neighborhood"])
houses.drop(houses[houses['Estimate; Total:'] >= 400].index, inplace = True)
houses.plot.bar(x='Neighborhood', y="Estimate; Total:", figsize=(15,5))
```

We can pretty much eliminate all of these neighborhoods completely from contention. They do not have enough households to sustain us as a delivery driver. Even if you were to order food once a week from doordash, which is a lot, the number of other deliverers in your area is going to make it impossible to live off of just these households. Additionally, we need to consider the population density. We want to deliver to areas where there are a lot of houses very close by, if there is only 150 total households in the North Shore for example, we won't be able to make multiple deliveries at one time. The houses would be too far apart for the food to stay warm, and then we might lose our tip. 

Let's look at the extreme other end now.

```python
houses = pd.read_csv('two.csv', usecols=["Estimate; Total:", "Neighborhood"])
houses.drop(houses[houses['Estimate; Total:'] <= 3000].index, inplace = True)
houses.plot.bar(x='Neighborhood', y="Estimate; Total:", figsize=(15,5))
```

These are the neighborhoods we want to work in. They have a high desnity of people and enough households that we will never not have work. Brookline, Shadyside, and Squirrel Hill South are distinct outliers in this group. Realistically those in the upper half of this dataset are all viable locations to work, but the three outliers will be taken into account for our final result.
